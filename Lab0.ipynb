## While Loop (less Pythonic)

```python
s = "RacecAr "

def is_palindrome1(string):
    clean_string = string.lower().replace(' ', '')
    first, last = 0, len(clean_string)-1
    while( first < last):
        if clean_string[first] == clean_string[last]:
            first +=1
            last -=1
        else:
            return False
    
    return True

is_palindrome1(s) 
```

## String Slicing (More Pythonic)

```python
def is_palindrome2(string):
    clean_string = string.lower().replace(' ', '')
    reversed_string = clean_string[::-1]
    return clean_string == reversed_string

s = " Rac eC ar "
is_palindrome2(s)
```

### Questions
1. How many movies have "Star Wars" in the movie_title field?

```python
import csv

num_sw_movie = 0
with open('movie.csv', encoding="utf-8") as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        movie_title = row['movie_title']
        if 'Star Wars' in movie_title:
            num_sw_movie += 1

num_sw_movie
```

2. Which movie has the highest IMDB Score?

```python
import csv

highest_imdb = 0
highest_movie = ''

with open('movie.csv') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        imdb_score = float(row['imdb_score'])
        if  highest_imdb < imdb_score:
            highest_imdb = imdb_score
            movie_title = row['movie_title']
            highest_movie = movie_title

highest_movie

```

3. Which movie has the lowest IMDB Score?


```python
import csv

lowest_imdb = 20
lowest_movie = ''

with open('movie.csv') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        imdb_score = float(row['imdb_score'])
        if  imdb_score < lowest_imdb:
            lowest_imdb = imdb_score
            movie_title = row['movie_title']
            lowest_movie = movie_title

lowest_movie
```

4. For each director, how many movies have they directed (from this dataset)?

```python
import csv

directors = {}

with open('movie.csv', encoding="utf-8") as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        director_name = row['director_name']
        if director_name != '':
            current_count = directors.get(row['director_name'],0)
            if current_count == 0:
                directors[row['director_name']] = 1
            else:
                directors[row['director_name']] += 1

directors
```

5. What's the average duration of Action movies?

```python

import csv

avg_duration = 0
add_duration = 0
number_rows = 0

with open('movie.csv') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        if 'Action' in row['genres']:
            if row['duration'] != '':
                duration = float(row['duration'])
                add_duration += duration
                number_rows += 1
    avg_duration = add_duration / number_rows
    # durations[row['duration']] = float(row['duration'])

avg_duration
```
